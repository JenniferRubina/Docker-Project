```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score,accuracy_score
```

```python
df=pd.read_csv('fraud_detection_dataset.csv')
df.head(10)
```

```python
df.columns
```

```python
df.drop(['Transaction ID','Customer ID','IP Address','Transaction Currency','Transaction Frequency'], axis=1, inplace=True)
```

```python
df
```

```python
y=df['Fraudulent']
y
```

```python
df.drop('Fraudulent', axis=1, inplace=True)
df
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])

# Extract year, month, day, hour, and minute features
df['Year'] = df['Transaction Date'].dt.year
df['Month'] = df['Transaction Date'].dt.month
df['Day'] = df['Transaction Date'].dt.day
df['Hour'] = df['Transaction Date'].dt.hour
df['Minute'] = df['Transaction Date'].dt.minute
df.drop('Transaction Date', axis=1, inplace=True)
df
```

```python
df.info()
```

```python
x= pd.get_dummies(df, columns = ['Transaction Type','Customer Gender'])
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier(class_weight='balanced', n_estimators=100, random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```
